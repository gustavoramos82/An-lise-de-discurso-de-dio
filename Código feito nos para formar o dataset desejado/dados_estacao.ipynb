Neste notebook será pego os dados de acordo ca estação metereologica e pode servir para filtros de um serie historica de indicie meteorolico e relacionar com os dados do resevatorio observando alguma relação

```python
import basedosdados as bd
import pandas as pd
```

```python
estacao = bd.read_table(dataset_id='br_inmet_bdmep',
table_id='estacao',
billing_project_id="471765969692")
estacao.head()
```

```python
municipio = pd.read_csv('https://raw.githubusercontent.com/leogermani/estados-e-municipios-ibge/master/municipios.csv')
municipio.head()
```

```python
estado = pd.read_csv('https://raw.githubusercontent.com/leogermani/estados-e-municipios-ibge/master/estados.csv')
estado.head()
```

```python
estado.drop('NOME',axis=1,inplace=True)
estado.head()
```

```python
estado.rename(columns={'COD':'COD UF'},inplace=True)
estado.head()
```

```python
estado_e_municipio = municipio.merge(estado,on='COD UF')
estado_e_municipio.head()
```

```python
estado_e_municipio.drop('COD UF',axis=1,inplace=True)
estado_e_municipio.head()
```

```python
estado_e_municipio.rename(columns={'COD':'id_municipio'},inplace=True)
estado_e_municipio.head()
```

```python
estado_e_municipio['id_municipio'] = estado_e_municipio['id_municipio'].astype(str)
estado_e_municipio.info()
```

```python
estacao = estacao.merge(estado_e_municipio,on='id_municipio')
estacao.head()
```

```python
estacao.drop('id_municipio',axis=1,inplace=True)
estacao.rename(columns={'NOME':'Municipio','SIGLA':'Sigla estado'},inplace=True)
estacao.head()
```
