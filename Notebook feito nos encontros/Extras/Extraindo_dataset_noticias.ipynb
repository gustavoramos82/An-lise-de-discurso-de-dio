# Extração de Noticias do Google news

Neste notebook vamos extrair as noticias do googlew news atravé da blibioteca python que tem uma api que retira noticias da plataforma google, suando os sehuintes termos

- Barragem
- Barragem Pará
- Barragem Mato Grosso
- Barragem Bahia

```python
!pip install GoogleNews
```

```python
from GoogleNews import GoogleNews
import pandas as pd
```

## Pesquisando os termos

```python
googlenews = GoogleNews(start='01/01/2000',end='31/12/2023')
googlenews.set_lang('pt')
googlenews.search('Barragem Mineração')
```

```python
barragem = pd.DataFrame(googlenews.result())
barragem
```

```python
barragem.shape
```

```python
googlenews = GoogleNews(start='01/01/2000',end='31/12/2023')
googlenews.set_lang('pt')
googlenews.search('Barragem Mineração Pará')

barragem2 = pd.DataFrame(googlenews.result())
barragem2
```

```python
googlenews = GoogleNews(start='01/01/2000',end='31/12/2023')
googlenews.set_lang('pt')
googlenews.search('Barragem Mineração Mato Grosso')

barragem3 = pd.DataFrame(googlenews.result())
barragem3
```

```python
googlenews = GoogleNews(start='01/01/2000',end='31/12/2023')
googlenews.set_lang('pt')
googlenews.search('Barragem Mineração Bahia')

barragem4 = pd.DataFrame(googlenews.result())
barragem4
```

```python
barragem.drop('img',axis=1,inplace=True)
barragem2.drop('img',axis=1,inplace=True)
barragem3.drop('img',axis=1,inplace=True)
barragem4.drop('img',axis=1,inplace=True)
```

```python
barragem
```

```python
lista_bar_1 = []

for i in range(barragem.shape[0]):
  lista_bar_1.append('Barragem Mineração')

```

```python
lista_bar_1
```

```python
barragem['Termo utilizado'] = lista_bar_1
barragem
```

```python
lista_bar_2 = []

for i in range(barragem2.shape[0]):
  lista_bar_2.append('Barragem Mineração Pará')

barragem2['Termo utilizado'] = lista_bar_2
barragem2
```

```python
lista_bar_3 = []

for i in range(barragem3.shape[0]):
  lista_bar_3.append('Barragem Mineração Mato Grosso')

barragem3['Termo utilizado'] = lista_bar_3
barragem3
```

```python
lista_bar_4 = []

for i in range(barragem4.shape[0]):
  lista_bar_4.append('Barragem Mineração Bahia')

barragem4['Termo utilizado'] = lista_bar_4
barragem4
```

```python
barragem_geral = pd.concat([barragem,barragem2,barragem3,barragem4])
barragem_geral
```

```python
barragem_geral.shape
```

```python
barragem_geral.columns = ['Titulo da noticia','Onde foi Tirado a noticias','Data','Hora',
                          'Descrição da noticia','Link da noticia','termo utilizado para pesquisa']

barragem_geral.head()
```

```python
barragem_geral.info()
```

```python
barragem_geral.drop('Hora',axis=1,inplace=True) # retirando essa coluna pois todos esses valores estão em branco
barragem_geral.head()
```

```python
barragem_geral['Onde foi Tirado a noticias'].value_counts()
```

```python
barragem_geral.to_csv('titulos tirado do google news.csv')
```
